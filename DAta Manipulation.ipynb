```python
## Data Agrregating and Grouping
df.head()
```

```python
grouped_mean=df.groupby('LandContour')['MSSubClass'].mean()
print(grouped_mean)
```

```python
## Merging and joining dataframe
## Create sample dataframe
df1=pd.DataFrame({'Key': ['A','B','C'],'Value1':[1,2,3]})
df2=pd.DataFrame({'Key': ['A','B','D'],'Value2':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
## Merge the dataframe on the "key Column"
pd.merge(df1,df2,on="Key",how="inner")
```

```python
pd.merge(df1,df2,on="Key",how="outer")
```

```python
pd.merge(df1,df2,on="Key",how="left") ## means more priority
```

```python
pd.merge(df1,df2,on="Key",how="right") 
```
